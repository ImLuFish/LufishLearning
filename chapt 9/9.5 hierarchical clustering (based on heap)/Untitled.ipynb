```python
import pandas as pd
```

```python
df = pd.read_csv("melon.csv")
```

```python
df.id = df.id + 1
```

```python
attrs = ["density", "sugar"]
```

```python
df
```

```python
def calc(x1, x2):
    res = 0
    for attr in attrs:
        res += (x1[attr] - x2[attr]) ** 2
    return res ** 0.5 
```

```python
tot = []
for i in range(len(df)-1):
    for j in range(len(df)-i):
        j += i
        if i != j:
            tot.append([i, j, calc(df.iloc[i], df.iloc[j])])
```

```python
tot = pd.DataFrame(tot)
```

```python
tot.columns = ["1", "2", "dist"]
```

```python
for i in range(len(tot)):
    tot.loc[i, ["id"]] = i+1
```

```python
tot1 = tot.copy()
```

```python
def init_heap():
    length = len(tot)
    for i in range(length):
        current_node = length - i
        update_heap(current_node)
```

```python
def swap(current_node, parent_node):
    tmp = tot.loc[parent_node-1, "dist"]
    tot.loc[parent_node-1, "dist"] = tot.loc[current_node-1, "dist"]
    tot.loc[current_node-1, "dist"] = tmp
    
def update_heap(current_node):
    print("current node is ", current_node) 
    if current_node == 1: 
        return
    parent_node = current_node // 2
    if tot.loc[parent_node-1, "dist"] > tot.loc[current_node-1, "dist"]:
        print("swap with ", parent_node)
        swap(current_node, parent_node)
        update_heap(parent_node)
```

```python
def min(x, y):
    if x < y:
        return 1
    return 0

def down_heap(current_node):
    print("current_node is ", current_node)
    if current_node * 2 > len(tot):
        print("quit")
        return
    left = tot.loc[current_node * 2 - 1, "dist"]
    right = 999
    if current_node * 2 + 1 <= len(tot):
        right = tot.loc[current_node * 2 + 1 - 1, "dist"]
    
    print(left, " ", right)
    if min(left, right):
        swap(current_node, current_node * 2)
        print("down left side")
        down_heap(current_node * 2)
    else:
        swap(current_node, current_node * 2 + 1)
        print("down right side")
        down_heap(current_node * 2 + 1)
```
