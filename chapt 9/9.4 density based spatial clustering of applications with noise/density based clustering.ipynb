```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("melon.csv")
```

```python
e = 0.11
minpts = 5 
```

```python
attrs = ["density", "sugar"]
```

```python
def calc(x1, x2):
    global attrs
    res = 0
    for attr in attrs:
        res += (x1[attr] - x2[attr]) ** 2
    return res ** 0.5
```

```python
for i in range(len(df)):
    for j in range(len(df)-i):
        j += i
        df.loc[i, str(j)] = calc(df.iloc[i], df.iloc[j])
```

```python
for i in range(len(df)):
    for j in range(i):
        df.loc[i, str(j)] = df.loc[j, str(i)]
```

```python
cols = [str(i) for i in range(len(df))]
```

```python
df.loc[:, cols] = df.loc[:, cols].where(df.loc[:, cols] <= 0.11)
```

```python
for i in range(len(df)):
    df.loc[i, "num"] = df.loc[i, cols].count() 
```
